```python

from sklearn.preprocessing import LabelEncoder #to find the correlation between the string and Integer
data={'course':['difficult','hard','moderate','easy','very easy'],
    'score':[10,20,30,40,50]}
encoder=LabelEncoder()
df=pd.DataFrame(data)
df['Difficulty_encoded']=encoder.fit_transform(df['course'])
print(df)
correlation=df['Difficulty_encoded'].corr(df['score'])
print("correlation:",correlation)
```

```python
data={'color':['red','blue','green','red'],
       'test_score':[50,60,70,80]}
df=pd.DataFrame(data)
df_encoded=pd.get_dummies(df,columns=['color'])
print(df_encoded)
correlation=df_encoded['color_red'].corr(df_encoded['test_score'])
print("correlation with red:",correlation)
```
