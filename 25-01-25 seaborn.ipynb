```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips') #to load the dataset
tips.head()
```

```python
tips
```

```python
#histogram for totalbill amounts
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="violet")
plt.title("Distribution of Total bills")
```

```python
#KDE PLOT
kernel density estimation(KDE) visualizes the probability density of a variable
```

```python
sns.kdeplot(data=tips,x="total_bill",fill=True,color="blue")  #shade == fill
plt.title("KDE plot of Total Bills")
```

```python
#BAR PLOT
bar plot display the mean of a quantitative variable for different categories
```

```python
#Average total bill by day
sns.barplot(data=tips,hue="day",y="total_bill",palette="viridis")
plt.title("Average total bill by day")
```

```python
#Box plot
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="coolwarm")
plt.title("Box Plot of Total bills by day and gender")
```

```python
#Scatter Plot
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("Scatter Plot of Total Bill vs. Tip")
```

```python
sns.lineplot(data=tips,x="size",y="tip",marker="o",color="green")
plt.title("Line plot of party size vs. Tip")
```

```python
sns.pairplot(data=tips,hue="sex",palette="husl")
```

```python
Heatmaps display data as a color-coded matrix
#correlation heatmap
```

```python
numeric_cols=tips.select_dtypes(include=['number']).columns
correlation_matrix=tips[numeric_cols].corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
```

```python
Task Question : What is the correlation between different numerical features in the dataset?
compute the correlation matrix and visualize it using the heat map
```

```python
A Heatmap of the correlation matrix allows us to see the strength and direction of relationship between the numeric variables.
This can highlight which features are strongly corelated,aiding in feature selection and understanding the data dyanmics.
```

```python
numeric_cols=df.select_dtypes(include=['number']).columns  # select only numeric columns
correlation_matrix=df[numeric_cols].corr() #compute the correlation matrix
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm") #plot the heat map
plt.title("Correlation Heatmap")
```
